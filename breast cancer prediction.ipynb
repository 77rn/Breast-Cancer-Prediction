```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import warnings

# Ignore all warnings
warnings.filterwarnings("ignore", category=Warning)
```

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
data_frame.shape
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model with increased max_iter
model = LogisticRegression(max_iter=10000)  # Increase max_iter to 1000 or higher

# Train the model
model.fit(X_train, Y_train)
```

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
Y_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, Y_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

```python
Y_train_binary = np.where(Y_train == 'malignant', 0, 1)  # Example conversion
Y_test_binary = np.where(Y_test == 'malignant', 0, 1)  # Example conversion
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train a Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train_binary)  # Use binary target for training
```

```python
from sklearn.metrics import accuracy_score

# Predict probabilities for the positive class (malignant)
Y_pred_rf = rf_model.predict(X_test)
accuracy_rf = accuracy_score(Y_test_binary, Y_pred_rf)
```

```python
print(f"Random Forest Classifier Accuracy: {accuracy_rf:.2f}")
```

```python
from sklearn.svm import SVC

# Create and train an SVM Classifier
svm_model = SVC(kernel='rbf', random_state=42)
svm_model.fit(X_train, Y_train)

# Make predictions on test data using SVM
Y_pred_svm = svm_model.predict(X_test)

# Calculate accuracy score for SVM
accuracy_svm = accuracy_score(Y_test, Y_pred_svm)
print(f"SVM Classifier Accuracy: {accuracy_svm:.2f}")
```
